```python
geometries = ["CHD6", "NDBR6", "NDBR50"]
result1 = "MOO_results_botorch/DP1000_400C_SwiftVoce_curve1"
result2 = "MOO_results_weighted_bo/DP1000_400C_SwiftVoce_curve1"
result_dest = "MOO_results/DP1000_400C_SwiftVoce_curve1"
```

```python
import shutil

for geometry in geometries:
    # Copy files from result1 to result_dest
    for simdata in os.listdir(f"{result1}/{geometry}/iteration/data"):
        src_dir = os.path.join(result1, geometry, "iteration/data", simdata)
        dest_dir = os.path.join(result_dest, geometry, "iteration/data", simdata)
        #print(src_dir)
        #print(dest_dir)
        shutil.copytree(src_dir, dest_dir)
    last_folder_number = len(os.listdir(os.path.join(result1, geometry, "iteration/data")))
    print(last_folder_number)
    
    for simdata in os.listdir(f"{result2}/{geometry}/iteration/data"):
        simDataNumber = int(simdata) + last_folder_number
        src_dir = os.path.join(result2, geometry, "iteration/data", simdata)
        dest_dir = os.path.join(result_dest, geometry, "iteration/data", str(simDataNumber))
        print(src_dir)
        print(dest_dir)
        shutil.copytree(src_dir, dest_dir)
```

```python
import shutil

for geometry in geometries:
    # Copy files from result1 to result_dest
    FD_Curves_smooth = {}
    FD_Curves_unsmooth = {}
    flowCurves = {}

    FD_Curves_smooth_result1 = np.load(f"{result1}/{geometry}/iteration/common/FD_Curves_smooth.npy", allow_pickle=True).tolist()
    FD_Curves_unsmooth_result1 = np.load(f"{result1}/{geometry}/iteration/common/FD_Curves_unsmooth.npy", allow_pickle=True).tolist()
    flowCurves_result1 = np.load(f"{result1}/{geometry}/iteration/common/flowCurves.npy", allow_pickle=True).tolist()

    FD_Curves_smooth_result2 = np.load(f"{result2}/{geometry}/iteration/common/FD_Curves_smooth.npy", allow_pickle=True).tolist()
    FD_Curves_unsmooth_result2 = np.load(f"{result2}/{geometry}/iteration/common/FD_Curves_unsmooth.npy", allow_pickle=True).tolist()
    flowCurves_result2 = np.load(f"{result2}/{geometry}/iteration/common/flowCurves.npy", allow_pickle=True).tolist()

    FD_Curves_smooth.update(FD_Curves_smooth_result1)
    FD_Curves_smooth.update(FD_Curves_smooth_result2)

    FD_Curves_unsmooth.update(FD_Curves_unsmooth_result1)
    FD_Curves_unsmooth.update(FD_Curves_unsmooth_result2)

    flowCurves.update(flowCurves_result1)
    print(len(flowCurves))
    flowCurves.update(flowCurves_result2)
    #print(len(FD_Curves_smooth))   
    #print(len(FD_Curves_unsmooth))
    print(len(flowCurves))

    # aaserting all keys are equal 
    for key in FD_Curves_smooth.keys(): 
        assert key in FD_Curves_unsmooth.keys() 
        assert key in flowCurves.keys()

    np.save(f"{result_dest}/{geometry}/iteration/common/FD_Curves_smooth.npy", FD_Curves_smooth)
    np.save(f"{result_dest}/{geometry}/iteration/common/FD_Curves_unsmooth.npy", FD_Curves_unsmooth)
    np.save(f"{result_dest}/{geometry}/iteration/common/flowCurves.npy", flowCurves)
```
