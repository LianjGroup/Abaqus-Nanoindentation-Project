```python
def read_PH_Curve(filePath):
    output_data=np.loadtxt(filePath, skiprows=2)
    # column 1 is time step
    # column 2 is pressure
    # column 3 is height
    columns=['X', 'Height', 'Pressure']
    df = pd.DataFrame(data=output_data, columns=columns)
    # Converting to numpy array
    height = df.iloc[:, 1].to_numpy()
    pressure = df.iloc[:, 2].to_numpy()
 
    return height, pressure
```

```python
height, pressure = read_PH_Curve("PH_Curve.txt")

# Plotting the PH curve
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(height, pressure)
plt.title('Pressure-Height Curve')
plt.ylabel('Pressure (N)')
plt.xlabel('Height (m)')
plt.show()
```
